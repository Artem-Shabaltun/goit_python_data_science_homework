- Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = pd.DataFrame(devs.loc[devs['Мова програмування'] == 'Python'])

python_data
```

- Визначте розмір таблиці python_data за допомогою методу shape

```python
python_data.shape
```

- Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data = python_data.groupby('Посада')
```

- Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
new_df = python_data['Зарплата на місяць'].agg(['min', 'max'])
new_df
```

- Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(data):
    return (data['min'] + data['max'])/2

avg = new_df.apply(fill_avg_salary, axis=1)
avg.name = 'avg'
new_df = pd.concat([new_df, avg], axis=1)
new_df
```

- Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
new_df['avg'].describe()
```

- Збережіть отриману таблицю в CSV файл

```python
new_df.to_csv('AS_2017_jun_final.csv')
```

- додайте від 3 до 5 графіків функцій різного типу на ваш вибір. Задайте графікам оформлення, щоб кожен графік у своїй домашній роботі чимось відрізнявся і не був схожим на інші. Можна використовувати як matplotlib, так і seaborn.

```python
languages = devs['Мова програмування'].value_counts(normalize=True)

other_languages = languages[languages < 0.02].sum()
languages = languages[languages >= 0.02]
languages['Інші'] = other_languages

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(languages.values, labels=languages.index, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Популярність мов програмування')
plt.show()
```

```python
grouped = devs.groupby('Мова програмування')['Зарплата на місяць'].mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(grouped['Мова програмування'], grouped['Зарплата на місяць'], color='green')

ax.set_xlabel('Мова програмування', fontsize=15, color='blue')
ax.set_xticklabels(grouped['Мова програмування'], rotation=45)

ax.set_ylabel('Середня зарплата на місяць', fontsize=15, color='blue')

ax.set_title('Середні заробітні плати спеціалістів різних мов програмування', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.violinplot(data=devs, x='Освіта', y='Зарплата на місяць')

plt.ylim(0, 10000)

plt.xlabel('Освіта', fontsize=15)
plt.ylabel('Зарплата на місяць', fontsize=15)
plt.title('Розподіл зарплат за рівнями освіти', fontsize=15)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()



```
